```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
#FIRST STEP:ADD A COLUMN WITH WEIGHTED RECENCY TO THE OPENED DATASET
#in order to weight the recency I tried to use the Fibonacci's sequence adding the difference the sequence: 
#1-2 -> 0,1,  2-3 -> +0,1, 3-5 -> +0,2, 5-8 -> 0,3, 8-13 -> 0,5, 13-21 -> 0,8, 21-34 -> 1,3

directory = os.listdir('/home/jovyan/A_Dataset_Graph/pipeline_1_output')
sorted_dir= sorted(directory)

for output_1 in sorted_dir:
    if output_1 not in ('.ipynb_checkpoints','.ds_store'):
        year_month_ds= pd.read_csv('/home/jovyan/A_Dataset_Graph/pipeline_1_output/'+output_1, dtype={'year_month':object})

        recency_weighted = []
        for value in year_month_ds['recency']:
            if value > 1095:
                value_w = value * 2.5
                recency_weighted.append(value_w)
            elif value > 730:
                value_w = value * 2.4
                recency_weighted.append(value_w)
            elif value > 545:
                value_w = value * 2.2
                recency_weighted.append(value_w)
            elif value > 365:
                value_w = value * 2
                recency_weighted.append(value_w)
            elif value > 180:
                value_w = value * 1.8
                recency_weighted.append(value_w)
            elif value > 90:
                value_w = value * 1.6
                recency_weighted.append(value_w)
            elif value > 60:
                value_w = value * 1.4
                recency_weighted.append(value_w)
            elif value > 30:
                value_w = value * 1.2
                recency_weighted.append(value_w)
            elif value > 15:
                value_w = value * 1.1
                recency_weighted.append(value_w)
            elif value > 7:
                value_w = value
                recency_weighted.append(value_w)
            else:
               value_w = value
               recency_weighted.append(value_w)

        year_month_ds['recency_weighted'] = recency_weighted
        year_month_ds[['recency_weighted']] = year_month_ds[['recency_weighted']].astype(np.int64)

        new_order = [0,1,8,2,3,4,5,6,7]
        year_month_ds_reordered = year_month_ds[year_month_ds.columns[new_order]]

        print(output_1)
        print(year_month_ds_reordered.describe())
        year_month_ds_reordered.loc[:,'response'] == 1
        print('Response Categories sizes')
        print(year_month_ds_reordered.loc[:,'response'].value_counts())

        year_month_ds_reordered.to_csv('/home/jovyan/A_Dataset_Graph/pipeline_2_output/'+ output_1 +'_rw.csv', index=False)
        del year_month_ds
        del year_month_ds_reordered
    

#SECOND STEP: VISUAL ANALYSYS

directory = os.listdir('/home/jovyan/A_Dataset_Graph/pipeline_2_output')
sorted_dir= sorted(directory)

for filename in sorted_dir:
    if filename not in ('.ipynb_checkpoints','.ds_store'):
        year_month_ds= pd.read_csv('/home/jovyan/A_Dataset_Graph/pipeline_2_output/'+filename)

        year_month_ds_sel = year_month_ds.loc[:, ['recency', 'recency_weighted', 'frequency', 'monetary', 'active_period', 'avg_period', 'response']]   

        print('')
        print('Heatmap for ' + filename + 'Recency Frequency Monetary values correlation')
        print('')
        heatmap=sns.heatmap(year_month_ds_sel.iloc[:, 0:8].corr(), square=True, linecolor="lightgray", linewidths=1, annot=True )
        plt.show()
        del year_month_ds

directory = os.listdir('/home/jovyan/A_Dataset_Graph/pipeline_2_output')
sorted_dir= sorted(directory)

for filename in sorted_dir:
    if filename not in ('.ipynb_checkpoints','.ds_store'):
        year_month_ds= pd.read_csv('/home/jovyan/A_Dataset_Graph/pipeline_2_output/'+filename)

        year_month_ds_sel = year_month_ds.loc[:, ['recency', 'recency_weighted', 'frequency', 'monetary', 'response']]   

        print('')
        print('Plot for ' + filename + 'Recency Frequency frequency weighted Monetary values')
        print('')
        fig, axes = plt.subplots(1,5, figsize=(20,5))
        for i, feature in enumerate (list(year_month_ds_sel)):
            #sns.distplot(year_month_ds_sel[feature], ax=axes[i])
            #sns.displot(year_month_ds_sel[feature], ax=axes[i])
            sns.histplot(data=year_month_ds_sel[feature], ax=axes[i], kde=True)
        plt.show()
        plt.close()
        print('')
        print(year_month_ds.describe())
        print('')
        print('Response Categories sizes')
        print(year_month_ds.loc[:,'response'].value_counts())
        del year_month_ds


#THIRD STEP:ASSIGNING RFM SCORES

directory = os.listdir('/home/jovyan/A_Dataset_Graph/pipeline_2_output')
sorted_dir= sorted(directory)

for output_2 in sorted_dir:
        if output_2 not in ('.ipynb_checkpoints','.ds_store'):
            rfm_df = pd.read_csv('/home/jovyan/A_Dataset_Graph/pipeline_2_output/'+output_2, dtype={'year_month':object})

        #rfm_df= pd.read_csv('DataBase_Dataset_Graph/pipeline_2_output/date_selector_rfm_2016.12.csv_rw.csv', dtype={'year_month':object})
        #rfm_df_sel = rfm_df.loc[:, ['user_id', 'recency', 'recency_weighted', 'frequency', 'monetary']]                                     

        print()
        print()
        print('file: '+output_2)
        print()
        print(rfm_df.head())
        print()
        print('numero di righe: ')
        print(rfm_df.shape[0])
        print()
        print('numero di colonne: ')
        print(rfm_df.shape[1])

        s1 =np.full((1,rfm_df.shape[0]-(int(0.1*rfm_df.shape[0])+int(0.1*rfm_df.shape[0])
                                            +int(0.1*rfm_df.shape[0])+int(0.1*rfm_df.shape[0])
                                            +int(0.1*rfm_df.shape[0])+int(0.1*rfm_df.shape[0])
                                            +int(0.1*rfm_df.shape[0])+int(0.1*rfm_df.shape[0])
                                            +int(0.1*rfm_df.shape[0]))),1)
        s2 =np.full((1,int(0.1*rfm_df.shape[0])),2)
        s3 =np.full((1,int(0.1*rfm_df.shape[0])),3)
        s4 =np.full((1,int(0.1*rfm_df.shape[0])),4)
        s5 =np.full((1,int(0.1*rfm_df.shape[0])),5)
        s6 =np.full((1,int(0.1*rfm_df.shape[0])),6)
        s7 =np.full((1,int(0.1*rfm_df.shape[0])),7)
        s8 =np.full((1,int(0.1*rfm_df.shape[0])),8)
        s9 =np.full((1,int(0.1*rfm_df.shape[0])),9)
        s10 =np.full((1,int(0.1*rfm_df.shape[0])),10)

        score = np.hstack((s1,s2,s3,s4,s5,s6, s7, s8, s9, s10)).flatten()


        rfm_df_rated = rfm_df.sort_values(by=['recency_weighted'], ascending= False)
        rfm_df_rated['r_w_score'] = score


        rfm_df_rated = rfm_df_rated.sort_values(by=['frequency'], ascending= True)
        rfm_df_rated['f_score'] = score


        rfm_df_rated = rfm_df_rated.sort_values(by=['monetary'], ascending= True)
        rfm_df_rated['m_score'] = score

        rfm_df_rated
        rfm_df_rated.reset_index(drop=True)
        rfm_df_rated.to_csv('/home/jovyan/A_Dataset_Graph/pipeline_3_output/'+ output_2 +'_scored.csv', index=False)
    
    
#seaching for min and max values: 

        for value in range(1,11):
            selectionMask = rfm_df_rated.loc[:,'r_w_score']==value
            print ('max value of recency, recency_weighted and frequency for r_w_score'+ str(value))
            print(rfm_df_rated.loc[selectionMask, ['recency','recency_weighted', 'frequency']].max())
            print()
            print ('min value of recency, recency_weighted and frequency for r_w_score'+ str(value))
            print(rfm_df_rated.loc[selectionMask, ['recency', 'recency_weighted', 'frequency']].min())
            print('--------------------------')       

        for value in range(1,11):
            selectionMask = rfm_df_rated.loc[:,'f_score']==value
            print ('max value of frequency, weighted recency and monetary for f_score'+ str(value))
            print(rfm_df_rated.loc[selectionMask, ['frequency', 'recency_weighted', 'monetary']].max())
            print()
            print ('min value of frequency, weighted recency and monetary for f_score'+ str(value))
            print(rfm_df_rated.loc[selectionMask, ['frequency', 'recency_weighted','monetary']].min())
            print('--------------------------')
```
